## Import libs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import os
from pathlib import Path
import re

from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, f1_score, classification_report, confusion_matrix
from catboost import CatBoostClassifier
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score

import torch
import torchvision.models as models
import torchvision.transforms as transforms
from torchvision.models import ResNet18_Weights, EfficientNet_B0_Weights

import warnings
warnings.filterwarnings('ignore')

# Настройка стилей
plt.style.use('ggplot')
```

## Data preparing

```python
class FashionDataLoader:
    def __init__(self, csv_path, images_dir):
        self.csv_path = csv_path
        self.images_dir = images_dir
        self.data = None
        
    def load_and_preprocess_data(self):
        """Загрузка и предобработка данных"""
        print("Загрузка данных...")
        self.data = pd.read_csv(self.csv_path)
        
        # Базовая очистка
        required_cols = ['reviews_total', 'reviews_avg_rating']
        self.data = self.data.dropna(subset=required_cols)
        self.data = self.data.drop(columns=['similar_count','similar_asins'])
        
        self.data['reviews_total'] = pd.to_numeric(self.data['reviews_total'], errors='coerce')
        self.data['reviews_avg_rating'] = pd.to_numeric(self.data['reviews_avg_rating'], errors='coerce')
        
        self.data = self.data[
            (self.data['reviews_total'] > 0) & 
            (self.data['reviews_avg_rating'] >= 0) &
            (self.data['reviews_avg_rating'] <= 5)
        ]
        
        print(f"Размер после очистки: {self.data.shape}")
        return self.data
    
    def create_target_variable(self, top_percentile=0.2):
        """Создание целевой переменной"""
        print("Создание целевой переменной...")
        
        self.data['success_score'] = (
            np.log1p(self.data['reviews_total']) *
            self.data['reviews_avg_rating']
        )
        
        threshold = np.percentile(self.data['success_score'], 100 * (1 - top_percentile))
        self.data['target'] = (self.data['success_score'] >= threshold).astype(int)
        
        success_count = self.data['target'].sum()
        total_count = len(self.data)
        
        print(f"Успешных товаров: {success_count} ({success_count/total_count:.1%})")
        print(f"Пороговое значение: {threshold:.2f}")
        
        return self.data
```

```python
def analyze_target_distribution(data):
    """Анализ распределения целевой переменной"""
    if 'group' in data.columns:
        plt.figure(figsize=(12, 6))
        category_success = data.groupby('group')['target'].mean().sort_values(ascending=False)
        category_success.plot(kind='bar')
        plt.title('Доля успешных товаров по категориям')
        plt.ylabel('Доля успешных товаров')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

## Extract image embeddings

```python
def extract_image_embeddings(data, images_dir, model_type='resnet'):
    """Извлечение эмбеддингов из изображений """
    print(f"Извлечение эмбеддингов с помощью {model_type}...")
    
    # Настройка преобразований
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                           std=[0.229, 0.224, 0.225])
    ])
    
    # Выбор модели
    if model_type == 'resnet':
        model = models.resnet18(weights=ResNet18_Weights.IMAGENET1K_V1)
        model = torch.nn.Sequential(*(list(model.children())[:-1]))  # Убираем последний слой
    elif model_type == 'efficientnet':
        model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
        model = torch.nn.Sequential(*(list(model.children())[:-1]))
    else:
        raise ValueError("Поддерживаемые модели: 'resnet', 'efficientnet'")
    
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    model.eval()
    
    # Извлечение эмбеддингов
    embeddings = []
    successful_extractions = 0
    
    for idx, row in data.iterrows():
        try:
            product_id = row.get('asin', str(idx))
            img_path = os.path.join(images_dir, f"{product_id}.jpg")
            image = Image.open(img_path).convert('RGB')
            image = transform(image).unsqueeze(0).to(device)
            
            with torch.no_grad():
                embedding = model(image)
                embedding = embedding.squeeze().cpu().numpy()
            
            embeddings.append(embedding)
            successful_extractions += 1
            
        except Exception as e:
            # Для ResNet18 размер эмбеддинга 512, для EfficientNet 1280
            embedding_size = 512 if model_type == 'resnet' else 1280
            embeddings.append(np.zeros(embedding_size))
            print(f"Ошибка при обработке изображения {product_id}: {e}")
    
    # Создаем DataFrame
    embedding_cols = [f'img_embed_{i}' for i in range(len(embeddings[0]))]
    embeddings_df = pd.DataFrame(embeddings, columns=embedding_cols)
    embeddings_df.index = data.index
    
    print(f"Успешно извлечено эмбеддингов: {successful_extractions}/{len(data)}")
    return embeddings_df
```

## Tabular data preparing

```python
def clean_category_string(cat_str):
    """ Очищает строку категории от служебных символов и цифр """
    # Удаляем все блоки в квадратных скобках с цифрами: [1234]
    cleaned_str = re.sub(r'\[\d+\]', '', cat_str)
    # Удаляем одиночные буквы в начале (like 'I' in 'IBooks') и лишние пробелы
    cleaned_str = re.sub(r'^\w ', '', cleaned_str).strip()
    # Заменяем несколько пробелов на один
    cleaned_str = re.sub(r'\s+', ' ', cleaned_str)
    return cleaned_str

def extract_nlp_embeddings(df):

    model = SentenceTransformer('all-MiniLM-L6-v2')

    cleaned_categories = df['categories'].apply(clean_category_string)

    category_embeddings = model.encode(cleaned_categories.tolist(), #numpy массив формы (n_samples, 384)
                                      show_progress_bar=True, 
                                      convert_to_numpy=True)
    
    embedding_columns = [f'category_embed_{i}' for i in range(category_embeddings.shape[1])]
    embeddings_df = pd.DataFrame(category_embeddings, columns=embedding_columns)
    
    return embeddings_df
```

```python
def prepare_tabular_features(data):
    """Подготовка табличных признаков"""    
    tabular_data = data.copy()
    
    nlp_embeddings_df = extract_nlp_embeddings(tabular_data)
    
    # Числовые признаки
    numeric_features = ['salesrank']
    for feature in numeric_features:
        if feature in tabular_data.columns:
            if tabular_data[feature].isnull().any():
                tabular_data[feature] = tabular_data[feature].fillna(tabular_data[feature].median())
            scaler = StandardScaler()
            tabular_data[feature] = scaler.fit_transform(tabular_data[[feature]])
    
    # Категориальные признаки
    categorical_features = ['group']
    for feature in categorical_features:
        if feature in tabular_data.columns:
            if tabular_data[feature].isnull().any():
                tabular_data[feature] = tabular_data[feature].fillna('unknown')
            
            if tabular_data[feature].nunique() <= 20:
                dummies = pd.get_dummies(tabular_data[feature], prefix=feature, drop_first=True)
                tabular_data = pd.concat([tabular_data, dummies], axis=1)
                tabular_data = tabular_data.drop(feature, axis=1)
            else:
                le = LabelEncoder()
                tabular_data[feature] = le.fit_transform(tabular_data[feature].astype(str))
    
    columns_to_drop = ['asin', 'success_score', 'reviews_total', 'reviews_avg_rating', 'title','has_image','image_filename', 'categories' ] + [col for col in tabular_data.columns if col.startswith('Unnamed')]
    columns_to_keep = [col for col in tabular_data.columns if col not in columns_to_drop]
    
    tabular_features = tabular_data[columns_to_keep]
    
    if not tabular_features.index.equals(nlp_embeddings_df.index):
        nlp_embeddings_df.index = tabular_features.index
    
    tabular_features = pd.concat([tabular_features, nlp_embeddings_df], axis=1)
    
    return tabular_features
```

## Model construction

```python
def train_and_evaluate_models(X_tabular, X_images, y):
    """Обучение и оценка моделей с CatBoost"""
    print("Обучение и оценка моделей с CatBoost...")
    
    # Разделение на train/test
    X_tabular_train, X_tabular_test, X_images_train, X_images_test, y_train, y_test = train_test_split(
        X_tabular, X_images, y, test_size=0.2, random_state=42, stratify=y
    )
    
    # Параметры CatBoost для предотвращения переобучения
    cb_params = {
        'iterations': 1000,  # Больше итераций, но с ранней остановкой
        'learning_rate': 0.01,  # Меньшая скорость обучения
        'depth': 3,  # Ограничение глубины деревьев
        'l2_leaf_reg': 10,  # Сильная L2 регуляризация
        'random_strength': 1,  # Контроль случайности для борьбы с переобучением
        'border_count': 128,  # Качество разбиений
        'auto_class_weights': 'Balanced',  # Балансировка классов
        'random_seed': 42,
        'verbose': False,  # Убираем вывод обучения
        'early_stopping_rounds': 50,  # Ранняя остановка
        'use_best_model': True  # Использовать лучшую модель
    }

    # Усиленная регуляризация CatBoost для image-модели
    cb_params_images = {
        'iterations': 300,  # Еще меньше итераций
        'learning_rate': 0.005,  # Еще меньше LR
        'depth': 2,  # Всего 2 уровня глубины
        'l2_leaf_reg': 20,  # Очень сильная регуляризация
        'random_strength': 5,  # Максимальная случайность
        'subsample': 0.6,  # Больше бэггинга
        'colsample_bylevel': 0.6,
        'auto_class_weights': 'Balanced',
        'early_stopping_rounds': 30,
        'use_best_model': True,
        'verbose': False
    }
    
    # Модель на табличных признаках
    print("\n1. CatBoost на табличных признаках:")
    model_tabular = CatBoostClassifier(**cb_params)
    model_tabular.fit(X_tabular_train, y_train, eval_set=(X_tabular_test, y_test))
    y_pred_tabular = model_tabular.predict_proba(X_tabular_test)[:, 1]
    
    # Модель на изображениях
    print("\n2. CatBoost на эмбеддингах изображений:")
    model_images = CatBoostClassifier(**cb_params_images)
    model_images.fit(X_images_train, y_train, eval_set=(X_images_test, y_test))
    y_pred_images = model_images.predict_proba(X_images_test)[:, 1]
    
    # Комбинированная модель
    print("\n3. CatBoost на комбинированных признаках:")
    X_combined_train = np.hstack([X_tabular_train, X_images_train])
    X_combined_test = np.hstack([X_tabular_test, X_images_test])
    
    model_combined = CatBoostClassifier(**cb_params)
    model_combined.fit(X_combined_train, y_train, eval_set=(X_combined_test, y_test))
    y_pred_combined = model_combined.predict_proba(X_combined_test)[:, 1]
    
    # Оценка моделей
    results = {}
    
    for name, y_pred, model in [
        ('Tabular', y_pred_tabular, model_tabular),
        ('Images', y_pred_images, model_images),
        ('Combined', y_pred_combined, model_combined)
    ]:
        
        auc = roc_auc_score(y_test, y_pred)
        f1 = f1_score(y_test, (y_pred > 0.5).astype(int))
        
        # Проверка на переобучение (разница между train и test)
        train_pred = model.predict_proba(X_tabular_train if name == 'Tabular' else 
                                       X_images_train if name == 'Images' else 
                                       X_combined_train)[:, 1]
        train_auc = roc_auc_score(y_train, train_pred)
        
        overfitting_gap = train_auc - auc
        best_iteration = model.get_best_iteration() if hasattr(model, 'get_best_iteration') else cb_params['iterations']
        
        results[name] = {
            'ROC-AUC': auc, 
            'F1-Score': f1,
            'Train-ROC-AUC': train_auc,
            'Overfitting-Gap': overfitting_gap,
            'Best-Iteration': best_iteration
        }
        
        print(f"{name}:")
        print(f"  Test ROC-AUC = {auc:.3f}, Train ROC-AUC = {train_auc:.3f}")
        print(f"  F1-Score = {f1:.3f}, Overfitting Gap = {overfitting_gap:.3f}")
        print(f"  Best Iteration = {best_iteration}")
        
        # Детальная отчетность для моделей с приемлемым переобучением
        if overfitting_gap < 0.1:
            y_pred_binary = (y_pred > 0.5).astype(int)
            print(f"  Classification Report:")
            print(classification_report(y_test, y_pred_binary))
    
    # Визуализация результатов
    _plot_model_comparison(results, y_test, [y_pred_tabular, y_pred_images, y_pred_combined], 
                          X_combined_train, y_train, X_combined_test)
    
    return results, model_combined, X_combined_test, y_test

def _plot_model_comparison(results, y_test, predictions, X_combined_train, y_train, X_combined_test):
    """Визуализация сравнения моделей"""
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # ROC-AUC сравнение
    model_names = list(results.keys())
    auc_scores = [results[name]['ROC-AUC'] for name in model_names]
    f1_scores = [results[name]['F1-Score'] for name in model_names]
    overfitting_gaps = [results[name]['Overfitting-Gap'] for name in model_names]
    
    ax1.bar(model_names, auc_scores, color=['skyblue', 'lightgreen', 'lightcoral'])
    ax1.set_title('ROC-AUC Scores by Model')
    ax1.set_ylabel('ROC-AUC Score')
    ax1.set_ylim(0, 1)
    
    ax2.bar(model_names, f1_scores, color=['skyblue', 'lightgreen', 'lightcoral'])
    ax2.set_title('F1 Scores by Model')
    ax2.set_ylabel('F1-Score')
    ax2.set_ylim(0, 1)
    
    
    # Confusion Matrix для комбинированной модели
    y_pred_combined_binary = (predictions[2] > 0.5).astype(int)
    cm = confusion_matrix(y_test, y_pred_combined_binary)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax3)
    ax3.set_title('Confusion Matrix - Combined Model')
    ax3.set_xlabel('Predicted')
    ax3.set_ylabel('Actual')
    
    # Learning curves (упрощенная версия)
    train_sizes = np.linspace(0.1, 1.0, 5)
    train_scores = []
    test_scores = []
    
    # Для комбинированной модели
    for size in train_sizes:
        n_samples = int(len(X_combined_train) * size)
        model = CatBoostClassifier(
            iterations=200,
            learning_rate=0.05,
            depth=3,
            random_seed=42,
            verbose=False
        )
        model.fit(X_combined_train[:n_samples], y_train[:n_samples])
        
        train_pred = model.predict_proba(X_combined_train[:n_samples])[:, 1]
        test_pred = model.predict_proba(X_combined_test)[:, 1]
        
        train_scores.append(roc_auc_score(y_train[:n_samples], train_pred))
        test_scores.append(roc_auc_score(y_test, test_pred))
    
    # Overfitting gap
    ax4.bar(model_names, overfitting_gaps, color=['orange', 'red', 'purple'])
    ax4.set_title('Overfitting Gap (Train AUC - Test AUC)')
    ax4.set_ylabel('Overfitting Gap')
    ax4.axhline(y=0.1, color='r', linestyle='--', alpha=0.7, label='Acceptable threshold')
    ax4.legend()
    
    plt.tight_layout()
    plt.show()

# Дополнительная функция для кросс-валидации
def cross_validate_cb(X, y, n_splits=5):
    """Кросс-валидация для CatBoost"""
    model = CatBoostClassifier(
        iterations=500,
        learning_rate=0.05,
        depth=4,
        l2_leaf_reg=3,
        auto_class_weights='Balanced',
        random_seed=42,
        verbose=False,
        early_stopping_rounds=30
    )
    
    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    cv_scores = cross_val_score(model, X, y, cv=cv, scoring='roc_auc', n_jobs=-1)
    
    print(f"Cross-Validation ROC-AUC Scores: {cv_scores}")
    print(f"Mean ROC-AUC: {cv_scores.mean():.3f} (±{cv_scores.std():.3f})")
    
    return cv_scores
```

## Clusterization and image analysis

```python
def cluster_image_embeddings(embeddings_df, data, target_column='target', n_clusters=None):
    """
    Кластеризация эмбеддингов изображений и анализ связи с успешностью товаров
    embeddings_df: DataFrame с эмбеддингами
    data: исходные данные с информацией о товарах
    n_clusters: количество кластеров (если None, подбирается автоматически)
    """
    
    # Объединяем эмбеддинги с исходными данными
    merged_data = data.copy()
    merged_data = pd.concat([merged_data, embeddings_df], axis=1)
    
    # Извлекаем матрицу эмбеддингов
    X = embeddings_df.values
    
    # Определяем оптимальное количество кластеров
    if n_clusters is None:
        n_clusters = find_optimal_clusters(X)
        print(f"Оптимальное количество кластеров: {n_clusters}")
    
    # Выполняем кластеризацию K-means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(X)
    
    # Добавляем информацию о кластерах к данным
    merged_data['image_cluster'] = clusters
    
    # Анализируем связь кластеров с успешностью
    cluster_analysis = analyze_clusters_success(merged_data, target_column)
    
    # Визуализируем результаты
    visualize_cluster_analysis(merged_data, cluster_analysis, target_column)
    
    return merged_data, kmeans, cluster_analysis

def find_optimal_clusters(X, max_clusters=10):
    """Поиск оптимального количества кластеров методом локтя и силуэта"""
    wcss = []  # Within-Cluster Sum of Square
    silhouette_scores = []
    
    for n in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=n, random_state=42, n_init=10)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)
        
        if len(np.unique(kmeans.labels_)) > 1:
            silhouette_scores.append(silhouette_score(X, kmeans.labels_))
        else:
            silhouette_scores.append(0)
    
    # Метод локтя 
    diffs = np.diff(wcss)
    diff_ratios = diffs[1:] / diffs[:-1]
    elbow_point = np.argmin(diff_ratios) + 3 
    
    # Метод силуэта 
    silhouette_point = np.argmax(silhouette_scores) + 2
    
    # Выбираем оптимальное значение
    optimal_clusters = min(elbow_point, silhouette_point)
    
    # Визуализация выбора количества кластеров
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    ax1.plot(range(2, max_clusters + 1), wcss, 'bo-')
    ax1.set_xlabel('Количество кластеров')
    ax1.set_ylabel('WCSS')
    ax1.set_title('Метод локтя')
    ax1.axvline(x=optimal_clusters, color='r', linestyle='--')
    
    ax2.plot(range(2, max_clusters + 1), silhouette_scores, 'go-')
    ax2.set_xlabel('Количество кластеров')
    ax2.set_ylabel('Silhouette Score')
    ax2.set_title('Метод силуэта')
    ax2.axvline(x=optimal_clusters, color='r', linestyle='--')
    
    plt.tight_layout()
    plt.show()
    
    return optimal_clusters

def analyze_clusters_success(data, target_column):
    """Анализ успешности товаров в разных кластерах"""
    cluster_stats = data.groupby('image_cluster').agg({
        target_column: ['mean', 'count', 'std']
    }).round(3)
    
    cluster_stats.columns = ['success_rate', 'count', 'std_dev']
    cluster_stats = cluster_stats.sort_values('success_rate', ascending=False)
    
    # Добавляем информацию о доле от общего количества
    total_count = len(data)
    cluster_stats['percentage'] = (cluster_stats['count'] / total_count * 100).round(1)
    
    print("Статистика успешности по кластерам:")
    print(cluster_stats)
    print(f"\nОбщее количество товаров: {total_count}")
    
    return cluster_stats

def visualize_cluster_analysis(data, cluster_stats, target_column):
    """Визуализация результатов кластерного анализа"""
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # График 1: Распределение успешности по кластерам
    success_by_cluster = data.groupby('image_cluster')[target_column].mean()
    ax1.bar(success_by_cluster.index, success_by_cluster.values)
    ax1.set_xlabel('Кластер')
    ax1.set_ylabel('Доля успешных товаров')
    ax1.set_title('Успешность по кластерам')
    ax1.grid(True, alpha=0.3)
    
    # График 2: Размер кластеров
    cluster_sizes = data['image_cluster'].value_counts().sort_index()
    ax2.bar(cluster_sizes.index, cluster_sizes.values)
    ax2.set_xlabel('Кластер')
    ax2.set_ylabel('Количество товаров')
    ax2.set_title('Размеры кластеров')
    ax2.grid(True, alpha=0.3)
    
    # График 3: Scatter plot успешности vs размера кластера
    ax3.scatter(cluster_stats['count'], cluster_stats['success_rate'], 
               s=100, alpha=0.7)
    ax3.set_xlabel('Размер кластера')
    ax3.set_ylabel('Доля успешных товаров')
    ax3.set_title('Успешность vs Размер кластера')
    ax3.grid(True, alpha=0.3)
    
    # Добавляем аннотации с номерами кластеров
    for i, (idx, row) in enumerate(cluster_stats.iterrows()):
        ax3.annotate(str(idx), (row['count'], row['success_rate']), 
                    xytext=(5, 5), textcoords='offset points')
    
    # График 4: Топ-5 самых успешных кластеров
    top_clusters = cluster_stats.nlargest(5, 'success_rate')
    ax4.bar(range(len(top_clusters)), top_clusters['success_rate'])
    ax4.set_xlabel('Кластер (топ-5)')
    ax4.set_ylabel('Доля успешных товаров')
    ax4.set_title('Топ-5 самых успешных кластеров')
    ax4.set_xticks(range(len(top_clusters)))
    ax4.set_xticklabels([f'Cluster {idx}\n(n={int(row["count"])})' 
                        for idx, row in top_clusters.iterrows()])
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Дополнительная информация
    print("\n" + "="*50)
    print("КЛЮЧЕВЫЕ ВЫВОДЫ:")
    print("="*50)
    
    # Самый успешный кластер
    best_cluster = cluster_stats.iloc[0]
    print(f"Самый успешный кластер: {cluster_stats.index[0]}")
    print(f"  - Доля успешных товаров: {best_cluster['success_rate']:.3f}")
    print(f"  - Количество товаров: {int(best_cluster['count'])}")
    print(f"  - Доля от общего числа: {best_cluster['percentage']}%")
    
    # Самый крупный кластер
    largest_cluster_idx = cluster_stats['count'].idxmax()
    largest_cluster = cluster_stats.loc[largest_cluster_idx]
    print(f"\nСамый крупный кластер: {largest_cluster_idx}")
    print(f"  - Доля успешных товаров: {largest_cluster['success_rate']:.3f}")
    print(f"  - Количество товаров: {int(largest_cluster['count'])}")
    print(f"  - Доля от общего числа: {largest_cluster['percentage']}%")
    
    # Разница между лучшим и худшим кластером
    worst_cluster = cluster_stats.iloc[-1]
    success_diff = best_cluster['success_rate'] - worst_cluster['success_rate']
    print(f"\nРазница в успешности между лучшим и худшим кластером: {success_diff:.3f}")
```

```python
def show_cluster_examples(data, images_dir, cluster_id, n_examples=5, target_column='target'):
    """Показать примеры товаров из определенного кластера"""
    cluster_data = data[data['image_cluster'] == cluster_id]
    
    if len(cluster_data) == 0:
        print(f"В кластере {cluster_id} нет товаров")
        return
    
    examples = cluster_data.sample(min(n_examples, len(cluster_data)))
    
    # Создаем подходящую сетку для отображения
    n_cols = min(5, len(examples))  # Максимум 5 колонок
    n_rows = (len(examples) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 3 * n_rows))
    
    # Если только одна строка или один столбец, преобразуем axes в 1D массив
    if n_rows == 1:
        axes = axes.reshape(1, -1)
    if n_cols == 1:
        axes = axes.reshape(-1, 1)
    
    # Скрываем лишние оси, если примеров меньше, чем ячеек
    for i in range(len(examples), n_rows * n_cols):
        row = i // n_cols
        col = i % n_cols
        axes[row, col].axis('off')
    
    for i, (idx, row) in enumerate(examples.iterrows()):
        row_idx = i // n_cols
        col_idx = i % n_cols
        
        try:
            product_id = row.get('asin', str(idx))
            img_path = os.path.join(images_dir, f"{product_id}.jpg")
            image = Image.open(img_path)
            
            axes[row_idx, col_idx].imshow(image)
            success_value = row.get(target_column, 'N/A')
            color = 'green' if success_value == 1 else 'red' if success_value == 0 else 'gray'
            axes[row_idx, col_idx].set_title(f"Cluster {cluster_id}\nSuccess: {success_value}", 
                                           color=color, fontweight='bold')
            axes[row_idx, col_idx].axis('off')
            
        except Exception as e:
            axes[row_idx, col_idx].text(0.5, 0.5, 'Image not found', 
                                      horizontalalignment='center', 
                                      verticalalignment='center',
                                      transform=axes[row_idx, col_idx].transAxes)
            axes[row_idx, col_idx].set_title(f"Cluster {cluster_id}\nSuccess: {row.get(target_column, 'N/A')}")
            axes[row_idx, col_idx].axis('off')
    
    plt.suptitle(f'Примеры товаров из кластера {cluster_id}\n(Зеленый = успешный, Красный = неуспешный)', 
                fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Выводим статистику кластера
    cluster_stats = data[data['image_cluster'] == cluster_id][target_column].value_counts()
    print(f"\nСтатистика кластера {cluster_id}:")
    print(f"   Всего товаров: {len(cluster_data)}")
    print(f"   Успешных (1): {cluster_stats.get(1, 0)}")
    print(f"   Неуспешных (0): {cluster_stats.get(0, 0)}")
    if len(cluster_data) > 0:
        success_rate = cluster_stats.get(1, 0) / len(cluster_data)
        print(f"   Доля успешных: {success_rate:.3f}")
```

```python
def visualize_clusters_with_tsne(X, clusters, y, target_column):
    """Визуализация кластеров с помощью t-SNE (более качественная)"""
    # Используем t-SNE для лучшей визуализации
    tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)  # Исправлено n_iter на max_iter
    X_tsne = tsne.fit_transform(X)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # График 1: Кластеры
    scatter1 = ax1.scatter(X_tsne[:, 0], X_tsne[:, 1], c=clusters, cmap='tab10', alpha=0.7)
    ax1.set_title('Кластеризация изображений (K-means) - t-SNE')
    ax1.set_xlabel('t-SNE Component 1')
    ax1.set_ylabel('t-SNE Component 2')
    plt.colorbar(scatter1, ax=ax1, label='Кластер')
    
    # График 2: Целевая переменная
    if y is not None:
        scatter2 = ax2.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='coolwarm', alpha=0.7)
        ax2.set_title(f'Распределение {target_column} - t-SNE')
        ax2.set_xlabel('t-SNE Component 1')
        ax2.set_ylabel('t-SNE Component 2')
        plt.colorbar(scatter2, ax=ax2, label=target_column)
    
    plt.tight_layout()
    plt.show()
```

## Pipeline

```python
def analysis():
    """Основной пайплайн анализа"""
    data_loader = FashionDataLoader(
        csv_path='final_dataset.csv',
        images_dir='amazon_images/'
    )
    
    # 1. Загрузка и подготовка данных
    data = data_loader.load_and_preprocess_data()
    
    # 2. Создание целевой переменной
    data = data_loader.create_target_variable(top_percentile=0.2)
    
    # 3. Анализ распределения
    analyze_target_distribution(data)
    
    # 4. Извлечение признаков
    embeddings_df = extract_image_embeddings(data, data_loader.images_dir, model_type='resnet')
    
    # Применяем PCA к эмбеддингам
    pca = PCA(n_components=50) 
    X_images_pca = pca.fit_transform(embeddings_df.values)
    
    # Создаем DataFrame из PCA-признаков
    X_images_pca_df = pd.DataFrame(X_images_pca, 
                                  columns=[f'img_pca_{i}' for i in range(X_images_pca.shape[1])],
                                  index=embeddings_df.index)
    
    X_tabular = prepare_tabular_features(data)
    
    # 5. Обучение моделей
    y = X_tabular['target']
    X_tabular = X_tabular.drop(columns='target')
    results, model, X_test, y_test = train_and_evaluate_models(
        X_tabular, X_images_pca, y
    )
    
    # 6. Кластеризация и анализ изображений
    # Используем оригинальные эмбеддинги (до PCA) для кластеризации
    merged_data, kmeans, cluster_analysis = cluster_image_embeddings(
        embeddings_df, data, 'target', n_clusters=5
    )
    
    # Получаем эмбеддинги и кластеры
    X_embeddings = embeddings_df.values  # Эмбеддинги изображений
    clusters = merged_data['image_cluster'].values  # Метки кластеров
    y_target = data['target'].values  # Целевая переменная
    
    # Вызываем функцию визуализации
    visualize_clusters_with_tsne(X_embeddings, clusters, y_target, 'target')
    
    # 7. Вывод результатов
    print("\n" + "="*50)
    print("ФИНАЛЬНЫЕ РЕЗУЛЬТАТЫ:")
    print("="*50)
    
    model_results = {k: v for k, v in results.items() if k != 'Combined_CV'}
    best_model = max(model_results.items(), key=lambda x: x[1]['ROC-AUC'])
    improvement = results['Combined']['ROC-AUC'] - results['Tabular']['ROC-AUC']
    
    print(f"Лучшая модель: {best_model[0]} (ROC-AUC: {best_model[1]['ROC-AUC']:.3f})")
    print(f"Улучшение от добавления изображений: {improvement:.3f}")
    
    if improvement > 0.05:
        print("Фото значительно улучшают прогноз успешности товара!")
    elif improvement > 0.02:
        print("Фото немного улучшают прогноз")
    else:
        print("Фото не оказывают значительного влияния на прогноз")
    
    return data, results, cluster_analysis, merged_data

data, results, cluster_analysis, merged_data = analysis()
```
